 ### array数组的数值计算

```python
import numpy as np
tang_array = np.array([[1,2,3],[4,5,6]])
tang_array
```

```python
np.sum(tang_array)
```

指定要进行的操作是沿着什么轴（维度）

```python
np.sum(tang_array,axis=0)
```

```python
tang_array.ndim
```

```python
np.sum(tang_array,axis=1)
```

```python
np.sum(tang_array,axis=-1)
```

高纬度数组想要沿着最后一个轴求和，可以用axis = -1

```python
tang_array.sum()
```

```python
tang_array
```

```python
tang_array.sum(axis = 0)
```

```python
tang_array.sum(axis = 1)
```

```python
tang_array.prod()
```

prod计算数组中所有元素求积

```python
tang_array.prod(axis = 0)
```

```python
tang_array.prod(axis = 1)
```

指定轴乘积


```python
tang_array.min()
```

```python
tang_array.min(axis = 0)
```

```python
tang_array.min(axis = 1)
```

横纵求最小值

```python
tang_array.max()
```

求最大值

找到索引位置

```python
tang_array.argmin()
```

找最小值在什么位置（索引）

```python
tang_array.argmin(axis = 0)
```

```python
tang_array.argmin(axis=1)
```

```python
tang_array.argmax()
```

```python
tang_array.mean()
```

均值

```python
tang_array.mean(axis = 0)
```
