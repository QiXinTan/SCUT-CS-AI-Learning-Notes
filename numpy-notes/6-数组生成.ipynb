```python
np.arange(2,20,2)
```

```python
np.linspace(0,10,10)
```

linspace双闭区间，构造n个数的等差数列，自动进行计算

不知道怎么用可以看帮助文档

```python
print(help(np.logspace(10,20)))
```

logspace默认是10为底的

```python
np.logspace(0,1,5)
```

第一个参数 0 (start)：这是指数的起始值。
代表从 $10^0$（即 1）开始。
第二个参数 1 (stop)：这是指数的结束值。
代表到 $10^1$（即 10）结束。
第三个参数 5 (num)：在这两个边界之间，一共生成 5 个数字。

第一步：系统先偷偷用 np.linspace(0, 1, 5) 生成 5 个均匀的指数：[0, 0.25, 0.5, 0.75, 1]。

第二步：把底数 10 的这些次方算出来：[$10^0$, $10^{0.25}$, $10^{0.5}$, $10^{0.75}$, $10^1$]。

```python
x = np.linspace(-10,10,5)
x
```

```python
y = np.linspace(-10,10,5)
y
```

```python
x, y= np.meshgrid(x,y)
```
