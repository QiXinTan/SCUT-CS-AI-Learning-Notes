### 数组形状操作

```python
import numpy as np
tang_array = np.arange(10)
tang_array
```

```python
tang_array.shape
```

```python
tang_array.shape = 2,5
```

将array的维度改为2维，2行5列

```python
tang_array
```

```python
tang_array.shape = 1,10
tang_array
```

```python
tang_array.reshape(1,10)
```

大小必须不能改变,数组n*m一定要等于元素个数

```python
tang_array .shape = 3,4
```

```python
tang_array = np.arange(10)
tang_array.shape
```

```python
tang_array = tang_array[np.newaxis,:]
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis,np.newaxis]
tang_array.shape
```

可以随意增加维度

```python
tang_array = np.arange(10)
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array.squeeze()
tang_array.shape
```

将空轴压缩去除

```python
tang_array.shape = 2,5
```

```python
tang_array
```

```python
tang_array.transpose()
```

转置，将2行5列转置为5行2列

```python
tang_array.T
```

transpose和.T转置效果相同，都不改变原数组，需要重新赋值

```python
tang_array
```

### 数组的连接

```python
a = np.array([[123,456,678],[3214,456,134]])
a
```

```python
b = np.array([[1235,3124,432],[43,13,134]])
b
```

a，b都是两行三列的

```python
c = np.concatenate((a,b))
c
```

concatenate进行连接操作，括号内需要将a，b
写成类似元组的形式（a，b）
默认按照第一个维度拼接（行拼接）

```python
c = np.concatenate((a,b),axis = 0)
c
```

```python
c = np.concatenate((a,b),axis = 1)
c
```

按照第二个维度拼接

```python
c.shape
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

vstack和hstack分别类似于行拼接、列拼接

```python
a
```

```python
a.flatten()
```

flatten按行拉平

```python
a
```

```python
a.ravel()
```

ravel也可以按行拉平
