### 使用Numpy读写数据

```python
import numpy as np
```

```python
%%writefile tang.txt
1 2 3 4 5 6
2 3 5 8 7 9
```

```python
data = []
with open('tang.txt') as f:
    for line in f.readlines():
        fileds = line.split()
        cur_data = [float(x) for x in fileds]
        data.append(cur_data)
data = np.array(data)
data
```

首先定义data作为数据容器
然后使用with open('tang.txt') as f: 打开文件
随后一次性读取文件并for循环一行一行处理
 fileds = line.split()把一行文本按空格切分
 随后将切分出来的字符串挨个青汁转换为float类型并存入打他中
 最后讲data转换为np.array结构

```python
data = np.loadtxt('tang.txt')
data
```

直接使用loadtxt

```python
%%writefile tang2.txt
1,2,3,4,5,6
2,3,5,8,7,9
```

```python
data = np.loadtxt('tang2.txt',delimiter = ',')
data
```

默认使用空格分隔符，如果使用“，”需要注明

```python
%%writefile tang2.txt
x,y,z,w,a,b
1,2,3,4,5,6
2,3,5,8,7,9
```

```python
data = np.loadtxt('tang2.txt',delimiter = ',',skiprows = 1)
data
```

* 'tang2.txt':路径最好放到和代码一起
* skiprows : 去掉几行
* delimiter = ',' :分隔符
* usecols = (0,1,4) ：指定使用哪几列 #最后的矩阵还可以选择使用哪几列

```python
tang_array = np.array([[1,2,3],[4,5,6]])
```

```python
np.savetxt('tang4.txt',tang_array)
```

将np数组直接写入文件

```python
np.savetxt('tang4.txt',tang_array,fmt='%d')
```

指定用整数保存，不然默认为浮点数

```python
np.savetxt('tang4.txt',tang_array,fmt='%d',delimiter = ',')
```

指定用整数且用逗号作为分隔符

```python
np.savetxt('tang4.txt',tang_array,fmt='%.2f',delimiter = ',')
```

指定保留小数位

### 读写array结构

```python
tang_array = np.array([[1,2,3],[4,5,6]])
np.save('tang_array.npy',tang_array)
```

```python
tang = np.load('tang_array.npy')
tang
```

.npy可以直接保存np.array的结构并直接使用load读取，不用转换

```python
tang_array2 = np.arange(10)
tang_array2
```

```python
np.savez('tang.npz',a=tang_array,b=tang_array2)
```

savez在npz中存储了a和b两个数组

```python
data = np.load('tang.npz')
```
