```python
tang_array = np.ones((5,5))
tang_array
```

```python
tang_array = np.pad(tang_array,pad_width = 1,mode = 'constant',constant_values = 0)
tang_array
```

我们来拆解这四个关键参数：

tang_array：你要处理的原数组。

pad_width = 2：填充的厚度。

意思是：在数组的上下左右（如果是二维）或者前后（如果是一维）各增加 2 层数据。

mode = 'constant'：填充模式。

constant 代表“常量填充”，也就是用一个固定的数字去填满新增加的那 2 层空间。

constant_values = 0：填充的具体数值。

既然选择了常量模式，这里就指定用 0 来填充。

```python
z1 = np.random.randint(0,10,10)
z2 = np.random.randint(0,10,10)
print (z1)
print (z2)
print (np.intersect1d(z1,z2))
#求两个数组的交集（即同时出现在 z1 和 z2 中的元素），且返回结果会自动去重 + 升序排序。
```

```python
# 找出z1中属于交集的元素的索引
intersect_vals = np.intersect1d(z1,z2)
indices = np.where(np.isin(z1, intersect_vals))[0]
print("z1中交集元素的索引：", indices)
```
