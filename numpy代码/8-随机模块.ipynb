### 随机的种子

```python
np.random.seed(101)
```

```python
mu, sigma = 0,0.1
np.random.normal(mu,sigma,10)
```

设定种子之后剩下所有随机都是依据种子来做的，当对比实验的时候，保证随机是“伪随机”，对比相同
