```python
tang_array = np.array([[1,0,6],
                       [1,7,0],
                       [2,3,1],
                       [2,4,0]])
tang_array
```

```python
index = np.lexsort([-1*tang_array[:,0],tang_array[:,2]])
index
```

分开指定排序
-1*（倒序）tang_array[:,0]（倒叙排列）

```python
tang_array = tang_array[index]
tang_array
```

使用索引排序
第三列升序（后排序但结果优先级更高），然后第一列降序（先排序）
