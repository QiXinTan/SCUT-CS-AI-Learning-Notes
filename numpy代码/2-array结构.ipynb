```python
tang_array = np.arange(0,100,10)
# 0 -100之间每10构造一个
tang_array
```

arrange按规则构造等差数列

```python
mask = np.array([9,0,0,1,7,8,0,0,1,1],dtype=bool)
#这里只有0，1也要指定bool类型，默认会存储为int64
mask
```

mask使用dtype = bool强制转换bool类型作为筛选条件

```python
tang_array[mask] #相当于筛选条件了，只有当mask对应位置为true的才保留在array中
```

使用数组嵌套数组的方法直接筛选

```python
random_array = np.random.rand(10)
random_array
```

```python
mask = random_array > 0.5 #直接利用表达式作为筛选条件
mask
```

```python
tang_array[mask]
```
